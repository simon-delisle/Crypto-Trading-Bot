```python
# Raw Package
import numpy as np
import pandas as pd

#Data Source
import yfinance as yf

#Data viz
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Custom classes
from Data_Preprocessing.technical_indicators import technical_analysis
ta = technical_analysis()
print(ta)
```

# Data

```python
data = yf.download(tickers='ETH-USD', period = '1y', interval = '1d')
data['Date'] = data.index
data
```

```python
df_candlestick_patterns = ta.find_candlestick_patterns(data)
df_candlestick_patterns
df_candlestick_patterns['candlestick_score'] = df_candlestick_patterns.iloc[:, 8:-1].sum(axis=1)
df_candlestick_patterns.to_csv('candlestick_patterns_test.csv')
```

```python
fig = ta.candlestick_chart(title='Ethereum Chart', 
                            time=data.Date, 
                            open=data.Open, 
                            high=data.High, 
                            low=data.Low, 
                            close=data.Close, 
                            volume=data.Volume, 
                            show_support=True, 
                            show_resistance=True,
                            show_local_mins=True, 
                            show_local_maxs=True,
                            weight_support_and_resistance=True,
                            weight_support_and_resistance_zones=200,
                            candlestick_score=df_candlestick_patterns['candlestick_score'].values
                        )
fig.show()
```
